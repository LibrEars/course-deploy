Accessing Remote Resources
==========================

I have mentioned before how one can access data files on your hard drive, but Python also allows you to access remote data, for example on the internet. The easiest way to do this is to use the built-in [urllib](http://docs.python.org/2/library/urllib.html) module. To start off, you have to open the URL and keep it as a Python object:

```python
from urllib import urlopen
u = urlopen('https://uebungen.physik.uni-heidelberg.de/uebungen/')
```

Note that in Python 3, you should instead use:

    from urllib.request import urlopen

Now, ``u`` will behave similarly to a (read-only) file object. You can read the contents as you would a normal file:

```python
content = u.read()
```

```python
print content[:300]  # only print the first 300 characters
```

You can either just use this information directly, or in some cases you might want to write it to a file. Let's download one of the full resolution files for the Ice coverage data from Problem Set 9:

```python
u2 = urlopen('http://mpia.de/~robitaille/share/ice_data/20060313.npy')
```

```python
content = u2.read()
```

Now we can open a new (binary) file and download the data to the file.

```python
f = open('20060313.npy', 'wb')
f.write(content)
f.close()
```
