## Daten einlesen

## Plotten

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10.0, 8.0) # TODO: get this workin
import matplotlib.pyplot as plt # Die Abkürzung `plt` ist Konvention
# Zeige Plots im Notebook an
%matplotlib inline
```

```python
l_closed, I_closed = np.loadtxt('resources/sonnenlicht_geschlossenes_fenster.txt', skiprows=2, unpack=True)
```

```python
plt.plot(l_open, I_open, label='offenes Fenster')
plt.plot(l_closed, I_closed, label='geschlossenes Fenster')
plt.title('Sonnenlichtspektrum')
plt.xlabel('Wellenlänge $\lambda \, [nm]$')
plt.ylabel('Intensität $I \, [\mathrm{counts}]$')
plt.legend()
```
